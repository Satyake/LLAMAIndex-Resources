```python
import os 
from dotenv import load_dotenv
from llama_index.core import Settings
from llama_index.llms.openai import OpenAI
from llama_index.core.agent.workflow import ReActAgent, AgentStream, ToolCallResult
from llama_index.core.workflow import Context
import asyncio


Settings.llm=OpenAI(model="gpt-4o-mini")

#def the tools
def multiply(a,b):
    """
    Docstring for multiply
    
    :param a: Description
    :param b: Description
    """
    return a*b
def add(a,b):
    """
    Docstring for add
    
    :param a: Description
    :param b: Description
    """
    return a+b

llm=Settings.llm

agent=ReActAgent(
llm=llm,
tools=[multiply, add]
)
context=Context(agent)

handler= agent.run(
    "what is the sum of 5 and 10, and then multiply the result",
    ctx=context
)
async for ev in handler.stream_events():
    if isinstance(ev, ToolCallResult):
        print(f"""Tool Call Result:{ev.tool_name} 
              returned {ev.tool_output}""")
    if isinstance(ev, AgentStream):
        print(ev.delta, end="", flush=True)

final_response=await handler
print(final_response)
    

```
