```python
import pandas as pd
import numpy as np

np.random.seed(42)

data = {
    "order_id": range(1001, 1021),
    "customer_id": np.random.choice([101,102,103,104,105], 20),
    "order_date": pd.date_range(start="2024-01-01", periods=20, freq="7D"),
    "country": np.random.choice(["Canada", "USA", "UK"], 20),
    "product_category": np.random.choice(["Electronics", "Clothing", "Home", "Sports"], 20),
    "quantity": np.random.randint(1, 5, 20),
    "price_per_unit": np.random.randint(20, 200, 20),
    "payment_method": np.random.choice(["Credit Card", "Debit Card", "PayPal"], 20),
    "discount_pct": np.random.choice([0, 5, 10, 15], 20),
    "returned": np.random.choice([0, 1], 20)
}

df = pd.DataFrame(data)

df["total_amount"] = df["quantity"] * df["price_per_unit"]
df["discount_amount"] = df["total_amount"] * df["discount_pct"] / 100
df["final_amount"] = df["total_amount"] - df["discount_amount"]

df.head()
```
